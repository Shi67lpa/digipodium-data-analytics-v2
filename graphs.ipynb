load the data analysis libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

get the data/generate data

```python
rainfall =[ 17,10,10,10,26,59]
months = ['jan','feb','mar','apr','may','june']
```

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

visualize the data

```python
plt.bar(months,rainfall,color='#ff8794')
```

```python
plt.pie(
    rainfall,
    labels=months,
    autopct='%1.1f%%',
)
plt.show()
```

```python
plt.pie(
    rainfall,
    labels=months,
    autopct='%1.1f%%',
    wedgeprops={'width':.6,'edgecolor':'#6688ff','linewidth':2},
    pctdistance=.8
)
plt.show()
```

box plot

```python
plt.boxplot(rainfall,
            vert= False,
            showmeans=True,
            meanline=True,
            meanprops={'color':'red'}
        )
plt.show()
```
